```python
def circle_space_points(number_points):
    point_list = []

    angle_space = np.pi/(number_points-1)
    print(angle_space)

    count = 0

    while count < number_points:

        current_point = (np.cos(angle_space*count),np.sin(angle_space*count))

        print(current_point)
        print(type(current_point))

        point_list.append(current_point)

        count += 1

    return point_list
```

```python
point_list = circle_space_points(8)
```

```python

from itertools import combinations


def combination_pair(point_list, amount):
    combination_list = combinations(point_list,amount)

    return combination_list

combined_list = combination_pair(point_list,2)

new_combined_list = []

for combination in combined_list:
    new_combined_list.append(combination)
    print(combination)

    for entry in combination:
        print(f"{combination} || {entry}")


```

```python
def center_of_mass(combination_pair):
    print(len(combination_pair))

    total = len(combination_pair)

    x_sum = 0
    y_sum = 0

    for pair in combination_pair:
        x_sum += pair[0]
        y_sum += pair[1]

    final_x = x_sum/total
    final_y = y_sum/total

    COM_point = (final_x,final_y)
    print(f"COM POINT: {COM_point}")

    return(COM_point)

COM_point_list = []

for entry in new_combined_list:
    print(entry)

    COM_point_list.append(center_of_mass(entry))

    print("BREAK")



combined_list_3 = combination_pair(point_list,3)

new_combined_list_3 = []

for combination in combined_list_3:
    new_combined_list_3.append(combination)
    print(combination)


COM_point_list_3 = []

for entry in new_combined_list_3:
    print(entry)

    COM_point_list_3.append(center_of_mass(entry))

    print("BREAK")




combined_list_4 = combination_pair(point_list,4)

new_combined_list_4 = []

for combination in combined_list_4:
    new_combined_list_4.append(combination)
    print(combination)


COM_point_list_4 = []

for entry in new_combined_list_4:
    print(entry)

    COM_point_list_4.append(center_of_mass(entry))

    print("BREAK")

```

```python
plt.figure(figsize=(8, 8))

count = 0

for point in point_list:
    count += 1
    print(f"Point: {count}")
    plt.scatter(point[0],point[1],color=(0,0,0),s=1)

for point in COM_point_list:
    count += 1
    print(f"Point: {count}")
    plt.scatter(point[0],point[1],color=(1,0,0),s=1)


for point in COM_point_list_3:
    count += 1
    print(f"Point: {count}")
    plt.scatter(point[0],point[1],color=(0,1,0),s=1)

for point in COM_point_list_4:
    count += 1
    print(f"Point: {count}")
    plt.scatter(point[0],point[1],color=(0,0,1),s=1)




plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```
