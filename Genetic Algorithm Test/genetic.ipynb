```python
def solve_species_solution(species_sequence):
    species_solution_list = []

    for entry in sequence_list:
        print(type(entry))
        entry_copy = np.copy(entry)
        print(f"Sequence {entry}")
        species_value = species_sequence*entry
        print(f"Species value {species_value}")
        species_solution_list.append(np.sum(species_sequence*entry))
    
    species_solution = np.array(species_solution_list)
    print(species_solution)
    return species_solution
```

```python
def fitness_value(species_solution,true_solution):
    count = 0

    error_sum = 0
    while count < len(true_solution):
        error = species_solution[count] - true_solution[count]
        square_error = error**2
        error_sum += square_error

        count+=1
    
    MSE = error_sum / len(true_solution)
    RMSE = np.sqrt(MSE)

    return RMSE
```

```python
species_solution = solve_species_solution(new_sequence)
print(fitness_value(species_solution,true_solution))
```

```python
# Create the population


population_count = 64

population_list = []

for count in range(population_count):
    population_list.append(np.random.choice([3, 0], size=80))

```

```python
# Compute one generation

fitness_list = []


for individual in population_list:
    species_solution = solve_species_solution(individual)
    fitness = fitness_value(species_solution,true_solution)
    fitness_list.append(fitness)
```

```python
paired_list = list(zip(fitness_list, population_list))
sorted_pair_list = sorted(paired_list, key=lambda x: x[0])

_, sorted_population_list = zip(*sorted_pair_list)
parent_list = sorted_population_list[:int(population_count/4)]
```

```python
def breed(parent_1,parent_2):
    recombination_point = random.randint(0,79)
    child_1 = np.concatenate((parent_1[:recombination_point],parent_2[recombination_point:]))
    child_2 = np.concatenate((parent_2[:recombination_point],parent_1[recombination_point:]))

    return child_1, child_2
```

```python
def mutate_individual(individual,mutation_chance):
    new_trait = []
    for base in individual:
        flip_try = random.random()
        if flip_try > mutation_chance:
            if base == 3:
                new_trait.append(0)
            else:
                new_trait.append(3)
        else:
            new_trait.append(base)
    np_new_trait = np.array(new_trait)

    return np_new_trait
```
