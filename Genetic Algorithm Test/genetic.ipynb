```python
import numpy as np
import copy
import random
import itertools
```

```python
sequence_list = []

sequence_1 = np.ones(80)
sequence_2 = np.concatenate((np.ones(40),np.zeros(40)))
sequence_3 = np.tile(np.concatenate((np.ones(20),np.zeros(20))),2)
sequence_4 = np.tile(np.concatenate((np.ones(10),np.zeros(10))),4)
sequence_5 = np.tile(np.concatenate((np.ones(5),np.zeros(5))),8)
sequence_6 = np.tile(np.concatenate((np.ones(2),np.zeros(2))),20)
sequence_7 = np.tile(np.array([1,0]),40)

sequence_list = [sequence_1,sequence_2,sequence_3,sequence_4,sequence_5,sequence_6,sequence_7]
true_solution = np.array([189,123,117,117,123,111,117])
```

```python
def solve_species_solution(species_sequence):
    species_solution_list = []

    for entry in sequence_list:
        species_value = species_sequence*entry
        species_solution_list.append(np.sum(species_sequence*entry))
    
    species_solution = np.array(species_solution_list)
    return species_solution
```

```python
def fitness_value(species_solution,true_solution):
    count = 0

    error_sum = 0
    while count < len(true_solution):
        error = species_solution[count] - true_solution[count]
        square_error = error**2
        error_sum += square_error

        count+=1
    
    MSE = error_sum / len(true_solution)
    RMSE = np.sqrt(MSE)

    return RMSE
```

```python
def breed(parent_1,parent_2):
    recombination_point = random.randint(0,79)
    child_1 = np.concatenate((parent_1[:recombination_point],parent_2[recombination_point:]))
    child_2 = np.concatenate((parent_2[:recombination_point],parent_1[recombination_point:]))

    return child_1, child_2
```

```python
def mutate_individual(individual,mutation_chance):
    new_trait = []
    for base in individual:
        flip_try = random.random()
        if flip_try > mutation_chance:
            if base == 3:
                new_trait.append(0)
            else:
                new_trait.append(3)
        else:
            new_trait.append(base)
    np_new_trait = np.array(new_trait)

    return np_new_trait
```

```python
# Create parent generation

population_count = 4096

parent_list = []

for count in range(population_count):
    parent_list.append(np.random.choice([3, 0], size=80))

# Set the parent generation as the next generation

next_gen = copy.copy(parent_list)
next_parent_count = 4

# Enter main loop here

generation_max = 100
current_gen = 0
mutation_rate = 0.1


while current_gen < generation_max:

    # Solve for fitness of each individual

    fitness_list = []

    for individual in next_gen:
        species_solution = solve_species_solution(individual)
        fitness = fitness_value(species_solution,true_solution)
        fitness_list.append(fitness)

    print(f"Current minimum RMSE: {min(fitness_list)}")

    # Select the top 16 individuals for parents
        

    paired_list = list(zip(fitness_list, next_gen))
    sorted_pair_list = sorted(paired_list, key=lambda x: x[0])


    _, sorted_population_list = zip(*sorted_pair_list)
    parent_list = sorted_population_list[:next_parent_count]

    # Create all parent combinations

    parent_combination_list = list(itertools.combinations(parent_list,2))
    combination_length = len(parent_combination_list)
    
    # Create empty new generation

    next_gen = []

    # Perform the breeding and mutation to populate next gen

    while len(next_gen) < population_count:
        random_pair_index = random.randint(0,combination_length-1)
        child_1, child_2 = breed(parent_combination_list[random_pair_index][0],parent_combination_list[random_pair_index][1])

        next_gen.append(mutate_individual(child_1,1-mutation_rate))
        next_gen.append(mutate_individual(child_2,1-mutation_rate))
    

    current_gen += 1
```

```python
fitness_list = []

for individual in next_gen:
    species_solution = solve_species_solution(individual)
    fitness = fitness_value(species_solution,true_solution)
    fitness_list.append(fitness)

paired_list = list(zip(fitness_list, next_gen))
sorted_pair_list = sorted(paired_list, key=lambda x: x[0])
_, sorted_population_list = zip(*sorted_pair_list)

best_fit_individual = sorted_population_list[0]

count = 0
for base in best_fit_individual:
    count += 1
    if base == 0:
        print(f"Prediction: {count}")
```
